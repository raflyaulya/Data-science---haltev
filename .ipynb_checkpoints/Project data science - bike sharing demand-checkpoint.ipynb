# Bike Sharing Demand Prediction in London

### Introduction
This project aims to predict the number of bike users in London based on certain parameters using machine learning techniques. The data source used for this project is train.csv and test.csv, which contain information about bike sharing demand in London. The data columns include datetime, season, holiday, workingday, weather, temp, atemp, humidity, windspeed, casual, registered, and count.

### Objectives
The main objective of this project is to build a machine learning model that can accurately predict the number of bike users in London based on various parameters. The model will be trained using the train.csv dataset and tested using the test.csv dataset. The ultimate goal is to achieve a high accuracy score on the test dataset.

### Methodology
This project will use various data science techniques to achieve the objective, including feature engineering, data visualization, and building several machine learning models such as linear regression, KNeighbors, decision tree, and gradient boosting regressor. The accuracy scores of each model will be evaluated using several methods such as r2score, RMSE, and cross-validation. If any of the models show signs of overfitting or underfitting, only the best model will be selected.

### Conclusion
In conclusion, this project aims to predict the number of bike users in London using machine learning techniques. The project will involve various data science techniques, and the ultimate goal is to achieve high accuracy on the test dataset using the best machine learning model.


## Data Train

```python
import pandas as pd

df = pd.read_csv('bike-sharing-demand/train.csv')
```

```python
df
```

```python
print('data shape:',df.shape)
print('data columns', df.columns)
```

```python
df.dtypes
```

```python
df
```

```python
import calendar
from datetime import datetime

df['date'] = df.datetime.apply(lambda x: x.split()[0])
df['hour'] = df.datetime.apply(lambda x: x.split()[1].split(':')[0])

df['weekday'] = df.date.apply(lambda x: calendar.day_name[datetime.strptime(x, '%Y-%m-%d').weekday()])
# df["month"] = df.date.apply(lambda a : calendar.month_name[datetime.strptime(a,"%Y-%m-%d").month])
df['month'] = df.date.apply(lambda x: calendar.month_name[datetime.strptime(x, '%Y-%m-%d').month])

df['season'] = df.season.map({1: 'Spring', 2 : 'Summer', 3 : 'Fall', 4 : 'Winter'})
# df['weather'] = df.weather.map({1: "Spring", 2 : "Summer", 3 : "Fall", 4 :"Winter" })
df["weather"] = df.weather.map({1: "Clear", 2 : "Mist", 3 : "Light Snow",4 :"Heavy Rain" })
```

```python
df.dtypes
```

```python
categoryVariableList = ["hour","weekday","month","season","weather","holiday","workingday"]
for var in categoryVariableList:
    df[var] = df[var].astype("category")
```

```python
df.dtypes
```

```python
df = df.drop(['datetime'], axis=1)
```

## Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sn

fig, axes = plt.subplots(nrows=2,ncols=2)
fig.set_size_inches(12, 10)
sn.boxplot(data=df,y="count",orient="v",ax=axes[0][0])
sn.boxplot(data=df,y="count",x="season",orient="v",ax=axes[0][1])
sn.boxplot(data=df,y="count",x="hour",orient="v",ax=axes[1][0])
sn.boxplot(data=df,y="count",x="workingday",orient="v",ax=axes[1][1])

axes[0][0].set(ylabel='Count',title="Box Plot On Count")
axes[0][1].set(xlabel='Season', ylabel='Count',title="Box Plot On Count Across Season")
axes[1][0].set(xlabel='Hour Of The Day', ylabel='Count',title="Box Plot On Count Across Hour Of The Day")
axes[1][1].set(xlabel='Working Day', ylabel='Count',title="Box Plot On Count Across Working Day")
```

```python
import numpy as np

corrMatt = df[["temp","atemp","casual","registered","humidity","windspeed","count"]].corr()
mask = np.array(corrMatt)
mask[np.tril_indices_from(mask)] = False
fig,ax= plt.subplots()
fig.set_size_inches(20,10)
sn.heatmap(corrMatt, mask=mask,vmax=.8, square=True,annot=True)
```

```python
fig,(ax1,ax2,ax3) = plt.subplots(ncols=3)
fig.set_size_inches(12, 5)
sn.regplot(x="temp", y="count", data=df,ax=ax1)
sn.regplot(x="windspeed", y="count", color='g', data=df,ax=ax2)
sn.regplot(x="humidity", y="count",color='r', data=df,ax=ax3)
```

```python
fig,(ax1,ax2,ax3,ax4,ax5)= plt.subplots(nrows=5)
fig.set_size_inches(12,20)
sortOrder = ["January","February","March","April","May","June","July","August","September","October","November","December"]
hueOrder = ["Sunday","Monday","Tuesday","Wednesday","Thursday","Friday","Saturday"]

monthAggregated = pd.DataFrame(df.groupby("month")["count"].mean()).reset_index()
monthSorted = monthAggregated.sort_values(by="count",ascending=False)
sn.barplot(data=monthSorted,x="month",y="count",ax=ax1,order=sortOrder)
ax1.set(xlabel='Month', ylabel='Avearage Count',title="Average Count By Month")

hourAggregated = pd.DataFrame(df.groupby(["hour","season"],sort=True)["count"].mean()).reset_index()
sn.pointplot(x=hourAggregated["hour"], y=hourAggregated["count"],hue=hourAggregated["season"], data=hourAggregated, join=True,ax=ax2)
ax2.set(xlabel='Hour Of The Day', ylabel='Users Count',title="Average Users Count By Hour Of The Day Across Season",label='big')

hourAggregated = pd.DataFrame(df.groupby(["hour","weekday"],sort=True)["count"].mean()).reset_index()
sn.pointplot(x=hourAggregated["hour"], y=hourAggregated["count"],hue=hourAggregated["weekday"],hue_order=hueOrder, data=hourAggregated, join=True,ax=ax3)
ax3.set(xlabel='Hour Of The Day', ylabel='Users Count',title="Average Users Count By Hour Of The Day Across Weekdays",label='big')

hourTransformed = pd.melt(df[["hour","casual","registered"]], id_vars=['hour'], value_vars=['casual', 'registered'])
hourAggregated = pd.DataFrame(hourTransformed.groupby(["hour","variable"],sort=True)["value"].mean()).reset_index()
sn.pointplot(x=hourAggregated["hour"], y=hourAggregated["value"],hue=hourAggregated["variable"],hue_order=["casual","registered"], data=hourAggregated, join=True,ax=ax4)
ax4.set(xlabel='Hour Of The Day', ylabel='Users Count',title="Average Users Count By Hour Of The Day Across User Type",label='big')

hourAggregated = pd.DataFrame(df.groupby(["hour","weather"],sort=True)["count"].mean()).reset_index()
sn.pointplot(x=hourAggregated["hour"], y=hourAggregated["count"],hue=hourAggregated["weather"], data=hourAggregated, join=True,ax=ax5)
ax5.set(xlabel='Hour Of The Day', ylabel='Users Count',title="Average Users Count By Hour Of The Day Across Weather",label='big')
```

## Feature Engineering

```python
exp = ['season', 'weather', 'weekday', 'month']

for i in exp:
    for elem in df[i].unique():
        df[str(elem)] = df[i] == elem

df.date = pd.to_datetime(df.date)
df.info()
```

```python
to_int = ['holiday', 'workingday', 'hour']

for x in to_int:
    df[x] = df[x].astype('int')
    
df.dtypes
```

## Linear Regression

```python
col = ['holiday', 'workingday', 'temp', 'atemp',
       'humidity', 'windspeed', 'casual', 'registered', #'date',
       'hour', 'Spring', 'Summer', 'Fall', 'Winter',
       'Clear', 'Mist', 'Light Snow', 'Heavy Rain', 'Saturday', 'Sunday',
       'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'January',
       'February', 'March', 'April', 'May', 'June', 'July', 'August',
       'September', 'October', 'November', 'December']

X = df[col].values
y = df['count'].values
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y , test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline

lr = LinearRegression(normalize=True)
lr = lr.fit(X_train, y_train)
```

```python
coef = lr.coef_
coef
```

```python
lr.intercept_
```

```python
lr.score(X_train, y_train)
```

```python
lr.score(X_test, y_test)
```

```python
import matplotlib.pyplot as plt

ypred_lr = lr.predict(X_test)
plt.scatter(y_test, ypred_lr)
```

## Evaluation metrics for Linear Regresion model

```python
from sklearn.model_selection import cross_val_score as cval

scores = cval(lr, X_train, y_train, cv=5)

print("Cross-validation scores: {}".format(scores))
print("Average cross-validation score: {:.2f}".format(scores.mean()))
print("Standard deviation of cross-validation scores: {:.2f}".format(scores.std()))
```

## Evaluation metrics for Data train

Evaluation metrics are used to determine the best model among several models that have been built. This is done by measuring the accuracy score using various metrics, including r2score, mean squared error (MSE), and root mean squared error (RMSE). These metrics provide a quantitative measure of the performance of each model, which can be used to compare and select the best model for the given problem. For these evaluation metrics, I am only comparing two models: the decision tree and gradient boosting regressor.

```python
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.model_selection import cross_val_score

acc_dt = r2_score(y_test, ypred_dt)
acc_gb = r2_score(y_test, ypred_gb)
rmse_dt = mean_squared_error(y_test, ypred_dt, squared=False)
rmse_gb = mean_squared_error(y_test, ypred_gb, squared=False)
cval_dt = cross_val_score(dt, X_train, y_train, cv=5, scoring='r2')
cval_gb = cross_val_score(gb, X_train, y_train, cv=5, scoring='r2')

print('R2 SCORE')
print('r2 score of Decision tree model:', acc_dt)
print('r2 score of Gradient boosting model:', acc_gb)

print('\nRMSE score')
print('RMSE decision tree score:', rmse_dt)
print('RMSE gradient boosting score:', rmse_gb)

print('\nCross-validation score')
print('Cross-validation decision tree:', cval_dt.mean())
print('Cross-validation gradient boosting:', cval_gb.mean())
```
