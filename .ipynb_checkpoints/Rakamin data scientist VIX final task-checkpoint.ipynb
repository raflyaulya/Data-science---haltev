# Modelling 

```python
df_col = df_train.columns
df_col = list(df_col)
col_remove = obj_columns + ['SK_ID_CURR', 'TARGET']

for x in col_remove:
    while x in df_col:
        df_col.remove(x)
        
# df_col
```

```python
# print(len(df_col))                235
# print(len(df_train.columns))      252

X = df_train[df_col]
y = df_train['TARGET']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y , test_size=0.3, random_state=101)
```

```python
from sklearn.feature_selection import SelectKBest, chi2

selected_feature = SelectKBest(score_func=chi2, k=10)
selected_feature.fit(X, y)
scores, columns = pd.DataFrame(selected_feature.scores_), pd.DataFrame(X.columns)
feature_scores = pd.concat([scores, columns], axis=1)
feature_scores.columns= ['Features', 'Scores']
feature_scores.sort_values(by=['Score'], ascending=False)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
ypred_lr = lr.predict(X_test)
```

```python
lr_train_score = lr.score(X_train, y_train)
print('Logistic Regression train score:', lr_train_score)
```

```python
lr_test_score = lr.score(X_test, y_test)
print('Logistic Regression test score:', lr_test_score)
```

## KNeighbors

```python
from sklearn import neighbors

kn = neighbors.KNeighborsClassifier(n_neighbors=9)
kn.fit(X_train, y_train)
```

```python
ypred_kn = kn.predict(X_test)
```

```python
kn_train_score = kn.score(X_train, y_train)
print('KNeighbors train score:', kn_train_score)
```

```python
kn_test_score = kn.score(X_test, y_test)
print('KNeighbors test score:', kn_test_score)
```
