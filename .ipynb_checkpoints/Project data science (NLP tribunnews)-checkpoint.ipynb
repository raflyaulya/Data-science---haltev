# Tribunnews: News Trend Analysis with NLP

This data science project aims to analyze news trends on a certain day from the Tribunnews.com website. The data used is taken from this source and processed using NLP (Natural Language Processing) techniques to identify the keywords that appear most often in the news on March 9, 2023.

Some of the libraries used in this project include Beautiful Soup, Literature for Indonesian, and NLTK (Natural Language Toolkit). These libraries help the process of scraping data from the Tribunnews.com website, cleaning the data that has been retrieved, and processing it to identify key words.

The method used in this data science project includes parsing HTML/XML from the target website source, cleaning data such as cleaning scripts and styles from parsed HTML, removing unnecessary numbers and symbols/characters, and tidying up text. After the data has been processed, this project also involves visualizing trends based on the keywords that appear most frequently in the news on March 9, 2023.

With this analysis of Tribunnews news trends, it is hoped that it can provide useful information for readers to understand what topics are trending on March 9, 2023 and help them make the right decisions in certain situations.

```python
import requests 
import string 
import re

from bs4 import BeautifulSoup
import nltk
from nltk.corpus import stopwords

!pip install Sastrawi
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
```

```python
web = requests.get('https://wartakota.tribunnews.com/').text
```

```python
type(web)
```

```python
soup = BeautifulSoup(web)
```

```python
for s in soup(['script', 'style']):
    s.decompose()     
```

```python
text = ' '.join(soup.stripped_strings)
```

```python
text = text.lower()
text = re.sub(r'\d+', '', text)
print(text)
```

```python
text = text.translate(str.maketrans('', '', string.punctuation))  
```

```python
factory = StemmerFactory()
stemmer = factory.create_stemmer()
output = stemmer.stem(text)
print(output)
```

```python
print(len(output))
```

```python
tokens = [word for word in output.split()]
print(tokens)
```

```python
nltk.download()
clean_tokens = tokens[:]
for token in tokens:
    if token in stopwords.words('indonesian'):
        clean_tokens.remove(token)
```

```python
clean_tokens
```

```python
freq = nltk.FreqDist(clean_tokens)
for key, val in freq.items():
    print(str(key),':',str(val))
```

```python
freq.plot(30)
```
