```python
import pymysql.cursors
```

```python
connection = pymysql.connect(host='localhost',
                            user='root',
                            database='netflix',
                            cursorclass=pymysql.cursors.DictCursor)
```

```python
cursor = connection.cursor()

# sql = 'SELECT country, COUNT(*) as user_cnt FROM user GROUP BY 1 ORDER BY 2 DESC'
sql2 = 'SELECT * FROM transaction'
cursor.execute(sql2)
transaction = cursor.fetchall()
# print(result)
# print(type(result))
```

```python
import pandas as pd

df2 = pd.DataFrame(transaction)
df2
```

```python
# df[['country', 'id']].groupby(['country']).count().reset_index().sort_values(['id'], ascending=False)

sql1 = 'SELECT * FROM user'
cursor.execute(sql1)
user = cursor.fetchall()
```

```python
df1 = pd.DataFrame(user)
df1
```

```python
df = df1.merge(df2, left_on='id', right_on='user_id')
df
```

```python
from collections import Counter

lis = []
for i in df['email']:
    lis.append(i)

count = Counter(lis)
mostCommon = count.most_common(1)[0][0]
print('most common email =',mostCommon)

lis2 = []
for i in df['first_name']:
    lis2.append(i)

count = Counter(lis2)
mostCommon = count.most_common(1)[0][0]
print('most common first name =',mostCommon)

print(lis)
print('\n ===========================================\n')
print(lis2)

# print(lis['Zulema'].count())
    

# print(lis.count())
# nama = lis.count()
# nama
```
