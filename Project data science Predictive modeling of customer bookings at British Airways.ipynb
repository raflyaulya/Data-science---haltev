### Exploratory data analysis

First, we must explore the data in order to better understand what we have and the statistical properties of the dataset.

To provide more context, below is a more detailed data description, explaining exactly what each column means:

- `num_passengers` = number of passengers travelling
- `sales_channel` = sales channel booking was made on
- `trip_type` = trip Type (Round Trip, One Way, Circle Trip)
- `purchase_lead` = number of days between travel date and booking date
- `length_of_stay` = number of days spent at destination
- `flight_hour` = hour of flight departure
- `flight_day` = day of week of flight departure
- `route` = origin -> destination flight route
- `booking_origin` = country from where booking was made
- `wants_extra_baggage` = if the customer wanted extra baggage in the booking
- `wants_preferred_seat` = if the customer wanted a preferred seat in the booking
- `wants_in_flight_meals` = if the customer wanted in-flight meals in the booking
- `flight_duration` = total duration of flight (in hours)
- `booking_complete` = flag indicating if the customer completed the booking

Before we compute any statistics on the data, lets do any necessary data conversion

# Data Preprocessing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve, roc_auc_score
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("data/customer_booking.csv", encoding="ISO-8859-1")
df.head()
```

```python
df.info()
```

```python
df["flight_day"].unique()
```

```python
mapping = {
    "Mon": 1,
    "Tue": 2,
    "Wed": 3,
    "Thu": 4,
    "Fri": 5,
    "Sat": 6,
    "Sun": 7,
}

df["flight_day"] = df["flight_day"].map(mapping)
```

```python
df["flight_day"].unique()
```

```python
df.describe()
```

```python
df
```

# Feature Engineering

```python
cat = ['sales_channel', 'trip_type',
       'booking_origin', 'route']

for var in cat:
    df[cat] = df[cat].astype('category')
    
```

```python
df.info()
```

```python
sns.pairplot(df)
plt.show()
```

```python
sns.heatmap(df.corr(), annot=True)
plt.show()
```

```python
df.isna().sum()
```

```python
lis = df['booking_origin']
lis.to_list()
```

```python
df2 = pd.DataFrame(lis)
```

```python
for elem in df['booking_origin'].unique():
    df2[str(elem)] = df['booking_origin'] == elem
```

```python
df2 = df2.drop(['booking_origin'], axis=1)
```

```python
df3 = df.join(df2)
```

```python
df3
```

```python
for elem in df['route'].unique():
    df2[str(elem)] = df['route'] == elem
    
```

```python
df3 = df.join(df2) 
```

```python
df3
```

```python
# EXPAND COLUMNS

df = df3

for elem in df['sales_channel'].unique():
    df[str(elem)] = df['sales_channel'] == elem
    
for elem in df['trip_type'].unique():
    df[str(elem)] = df['trip_type'] == elem    
```

```python
X = df.drop(['trip_type', 'sales_channel', 'route',
            'booking_origin', 'booking_complete'], axis=1)   

y = df['booking_complete']

X_train, X_test, y_train, y_test = train_test_split(X, y , test_size=0.3, random_state=101)
```

# Machine Learning

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
print('lr coef:', lr.coef_)
print()
print('lr intercept', lr.intercept_)
```

```python
ypred_lr = lr.predict(X_test)
```

```python
print('lr Train score:', lr.score(X_train, y_train))
print('lr Test score:', lr.score(X_test, y_test))
```

## KNeighbors

```python
from sklearn import neighbors

kn = neighbors.KNeighborsClassifier(n_neighbors=5)
kn.fit(X_train, y_train)
```

```python
ypred_kn = kn.predict(X_test)
```

```python
kn_train = kn.score(X_train, y_train)
print('KN train score:', kn_train)

kn_test = kn.score(X_test, y_test)
print('KN test score', kn_test)
```

## Decision Tree

```python
from sklearn import tree

dt = tree.DecisionTreeClassifier(max_depth=9)
dt.fit(X_train, y_train)
```

```python
ypred_dt = dt.predict(X_test)
```

```python
train_dt = dt.score(X_train, y_train)
print('decision tree train score:', train_dt)

test_dt = dt.score(X_test, y_test)
print('decision tree test score:', test_dt)
```

## Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier(max_depth=7)
gb.fit(X_train, y_train)
```

```python
train_gb = gb.score(X_train, y_train)
test_gb = gb.score(X_test, y_test)

print('Gradient boosting train score:', train_gb)
print('Gradient boosting test score:', test_gb)
```

```python
ypred_gb = gb.predict(X_test)
```

# Model Evaluation

```python
# Confusion Matrix 

cm_lr = confusion_matrix(y_test, ypred_lr)
cm_kn = confusion_matrix(y_test, ypred_kn)
cm_dt = confusion_matrix(y_test, ypred_dt)
cm_gb = confusion_matrix(y_test, ypred_gb)

print('confusion matrix logistic regression:\n',cm_lr)
print()
print('confusion matrix KNeighbors:\n',cm_kn)
print()
print('confusion matrix decision tree:\n',cm_dt)
print()
print('confusion matrix Gradient boosting:\n',cm_gb)
```

```python
# Define class labels
class_names = ['Positive', 'Negative']

# Make predictions for each model
y_pred_dt = dt.predict(X_test)
y_pred_knn = kn.predict(X_test)
y_pred_gb = gb.predict(X_test)
y_pred_lr = lr.predict(X_test)

# Compute confusion matrices
cm_dt = confusion_matrix(y_test, y_pred_dt)
cm_knn = confusion_matrix(y_test, y_pred_knn)
cm_gb = confusion_matrix(y_test, y_pred_gb)
cm_lr = confusion_matrix(y_test, y_pred_lr)

# Plot confusion matrices
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 12))

sns.heatmap(cm_lr, annot=True, cmap="Blues", fmt='d', xticklabels=class_names, yticklabels=class_names, ax=axes[0, 0])
axes[0, 0].set_title('Confusion Matrix for Logistic Regression')

sns.heatmap(cm_dt, annot=True, cmap="Blues", fmt='d', xticklabels=class_names, yticklabels=class_names, ax=axes[0, 1])
axes[0, 1].set_title('Confusion Matrix for Decision Tree')

sns.heatmap(cm_knn, annot=True, cmap="Blues", fmt='d', xticklabels=class_names, yticklabels=class_names, ax=axes[1,0])
axes[1, 0].set_title('Confusion Matrix for K-Nearest Neighbors')

sns.heatmap(cm_gb, annot=True, cmap="Blues", fmt='d', xticklabels=class_names, yticklabels=class_names, ax=axes[1, 1])
axes[1, 1].set_title('Confusion Matrix for Gradient Boosting')

for ax in axes.flat:
    ax.set(ylabel='True label', xlabel='Predicted label')

plt.show()
```

```python
recall_lr = cm_lr[0][0]/ (cm_lr[0][0] + cm_lr[0][1])
# recall_lr = str(recall_lr)
precision_lr = cm_lr[0][0] / (cm_lr[0][0] + cm_lr[1][0])
# precision_lr = str(precision_lr)
recall_kn = cm_kn[0][0] / (cm_kn[0][0] + cm_kn[0][1])
# recall_kn = str(recall_kn)
precision_kn = cm_kn[0][0] / (cm_kn[0][0] + cm_kn[1][0])
# precision_kn = str(precision_kn)
recall_dt = cm_dt[0][0]/ (cm_dt[0][0] + cm_dt[0][1])
# recall_dt = str(recall_dt)
precision_dt = cm_dt[0][0] / (cm_dt[0][0] + cm_dt[1][0])
# precision_dt = str(precision_dt)
recall_gb = cm_gb[0][0]/ (cm_gb[0][0] + cm_gb[0][1])
# recall_gb = str(recall_gb)
precision_gb = cm_gb[0][0] / (cm_gb[0][0] + cm_gb[1][0])
# precision_gb = str(precision_gb)

print('Logistic regression precision & recall')
print('recall logistic regression model:', recall_lr)
print('Precision logistic regression model:', precision_lr)

print('\nKNeighbors precision & recall')
print('recall KNeighbors model:', recall_kn)
print('Precision KNeighbros model:', precision_kn)

print('\nDecision tree precision & recall')
print('recall Decision tree model:', recall_dt)
print('Precision Decision tree model:', precision_dt)

print('\nGradient boosting precision & recall')
print('recall Gradient boosting model:', recall_gb)
print('Precision Gradient boosting model:', precision_gb)
```

```python
models = ['Logistic Regression', 'KNeighbors', 'Decision Tree', 'Gradient Boosting']
recall_scores = [recall_lr, recall_kn, recall_dt, recall_gb]
precision_scores = [precision_lr, precision_kn, precision_dt, precision_gb]

fig, ax = plt.subplots(figsize=(8, 5))

# Set the x and y axis labels
ax.set_xlabel('Models')
ax.set_ylabel('Score')

# the title
ax.set_title('Precision and Recall Scores')

# Create the bar plot
ax.bar(models, recall_scores, label='Recall')
ax.bar(models, precision_scores, label='Precision')

# Add the legend
ax.legend()

plt.show()
```

```python
models = ['Logistic Regression', 'KNeighbors', 'Decision Tree', 'Gradient Boosting']
recall_scores = [recall_lr, recall_kn, recall_dt, recall_gb]
precision_scores = [precision_lr, precision_kn, precision_dt, precision_gb]

# Create a numpy array of the precision and recall scores
scores = np.array([recall_scores, precision_scores])

# Create a heatmap of the precision and recall scores
ax = sns.heatmap(scores, annot=True, cmap='YlGnBu', xticklabels=models, yticklabels=['Recall', 'Precision'])

# Set the title
ax.set_title('Precision and Recall Scores')

plt.show()
```

```python
mse_dt = mean_squared_error(y_test, ypred_dt)
mse_kn = mean_squared_error(y_test, ypred_kn)
mse_gb = mean_squared_error(y_test, ypred_gb)

# calculate accuracy score
accuracy_dt = accuracy_score(y_test, ypred_dt)
accuracy_kn = accuracy_score(y_test, ypred_kn)
accuracy_gb = accuracy_score(y_test, ypred_gb)

# calculate cross validation score on the training data
cv_scores_dt = cross_val_score(dt, X_train, y_train, cv=5)
cv_scores_kn = cross_val_score(kn, X_train, y_train, cv=5)
cv_scores_gb = cross_val_score(gb, X_train, y_train, cv=5)

# print results
print("Decision tree Mean squared error: {}".format(mse_dt))
print("Decision tree Accuracy score: {}".format(accuracy_dt))
print("Decision tree Cross validation scores: {}".format(cv_scores_dt.mean()))
print('============================================\n')

print("KNeighbors Mean squared error: {}".format(mse_kn))
print("KNeighbors Accuracy score: {}".format(accuracy_kn))
print("KNeighbors Cross validation scores: {}".format(cv_scores_kn.mean()))
print('============================================\n')

print("Gradient boosting Mean squared error: {}".format(mse_gb))
print("Gradient boosting Accuracy score: {}".format(accuracy_gb))
print("Gradient boosting Cross validation scores: {}".format(cv_scores_gb.mean()))
```

```python
# store the results in lists
mse = [mse_dt, mse_kn, mse_gb]
accuracy = [accuracy_dt, accuracy_kn, accuracy_gb]
cv_scores = [cv_scores_dt.mean(), cv_scores_kn.mean(), cv_scores_gb.mean()]

# create the figure and axes
fig, ax = plt.subplots(figsize=(8,6))

# set the x-axis labels and the bar positions
labels = ['Decision tree', 'KNeighbors', 'Gradient boosting']
x_pos = np.arange(len(labels))

# plot the mean squared error scores
ax.bar(x_pos, mse, align='center', alpha=0.5, label='Mean squared error')

# plot the accuracy scores
ax.bar(x_pos, accuracy, align='center', alpha=0.5, label='Accuracy')

# plot the cross validation scores
ax.bar(x_pos, cv_scores, align='center', alpha=0.5, label='Cross validation')

# set the axis labels and title
ax.set_ylabel('Score')
ax.set_xticks(x_pos)
ax.set_xticklabels(labels)
ax.set_title('Model performance')

# add a legend
ax.legend(loc='center right')

# display the plot
plt.show()
```

```python
# plot bar chart
labels = ['Decision Tree', 'K-Neighbors', 'Gradient Boosting']
mse_scores = [mse_dt, mse_kn, mse_gb]
accuracy_scores = [accuracy_dt, accuracy_kn, accuracy_gb]
cv_scores = [cv_scores_dt.mean(), cv_scores_kn.mean(), cv_scores_gb.mean()]

x = np.arange(len(labels))
width = 0.2

fig, ax = plt.subplots()
rects1 = ax.bar(x - width, mse_scores, width, label='MSE')
rects2 = ax.bar(x, accuracy_scores, width, label='Accuracy')
rects3 = ax.bar(x + width, cv_scores, width, label='CV Scores')

# add value labels to each bar
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{:.3f}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom')
        
autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Model Performance Scores')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend(loc='lower right')
fig.tight_layout()

plt.show()
```

```python
pred_prob1 = dt.predict_proba(X_test)
pred_prob2 = gb.predict_proba(X_test)
pred_prob3 = kn.predict_proba(X_test)

fpr1, tpr1, thresholds1 = roc_curve(y_test, ypred_dt, pos_label=1)
fpr2, tpr2, thresholds2 = roc_curve(y_test, ypred_gb, pos_label=1)
fpr3, tpr3, thresholds3 = roc_curve(y_test, ypred_kn, pos_label=1)

random_probs = [0 for i in range(len(y_test))]
p_fpr, p_tpr, _ = roc_curve(y_test, random_probs, pos_label=1)

auc_score1 = roc_auc_score(y_test, pred_prob1[:,1])
auc_score2 = roc_auc_score(y_test, pred_prob2[:,1])
auc_score3 = roc_auc_score(y_test, pred_prob3[:,1])

print('Decision tree AUC score:',auc_score1)
print('Gradient boosting AUC score:',auc_score2)
print('KNeighbors AUC score:', auc_score3)
```

```python
roc_auc1 = roc_auc_score(y_test, pred_prob1[:, 1])  
roc_auc2 = roc_auc_score(y_test, pred_prob2[:,1], multi_class='ovr')  
roc_auc3 = roc_auc_score(y_test, pred_prob3[:, 1])  

plt.plot(fpr1, tpr1, color='darkorange', label='Decision tree AUC = %0.2f' % roc_auc1)  
plt.plot(fpr2, tpr2, color='red', label='Gradient boosting AUC = %0.2f' % roc_auc2) 
plt.plot(fpr3, tpr3, color='green', label='KNeighbors AUC = %0.2f' % roc_auc3)  

plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
plt.xlim([0.0, 1])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc="lower right")
plt.show()
```
