```python
import pandas as pd
import numpy as np

url = 'https://raw.githubusercontent.com/rizalespe/Dataset-Sentimen-Analisis-Bahasa-Indonesia/master/dataset_tweet_sentiment_cellular_service_provider.csv'
df = pd.read_csv(url)
```

```python
def text_processing(text):
    text = text.lower()
    text = re.sub(r'\d+', '', text)   # ngilangin angka
    text = text.translate(str.maketrans('', '', string.punctuation))  # ngilanging symbol dan character
    
    factory = StemmerFactory()
    stemmer = factory.create_stemmer()
    output = stemmer.stem(text)
    
    tokens = [word for word in output.split()]
    
    clean_tokens = tokens[:]
    for token in tokens:
        if token in stopwords.words('indonesian'):
            clean_tokens.remove(token)
            
    freq = nltk.FreqDist(clean_tokens)
    
    return freq
```

```python
df['tokens'] = df['Text Tweet'].apply(lambda x: text_processing(x))
```

```python
ser = df['tokens']
```

```python
ser.to_list()[0]
```

```python
def cond_freq_dist(lofd):
    lis = []
    for fd in lofd:
        dic = {}
        for key, val in fd.items():
            dic[key] = val
        lis.append(dic)
        
    return lis
```

```python
lis = cond_freq_dist(ser.to_list())
```

```python
df2 = pd.DataFrame(lis)
```

```python
df2.fillna(0, inplace=True)
```

```python
df2
```

```python
df3 = df.join(df2)
```

```python
df3
```

```python
df3['positive'] = df3['Sentiment'].apply(lambda x: 1 if x== 'positive' else 0) 
X = df3.drop(['Id', 'Sentiment', 'Text Tweet', 'tokens', 'positive'], axis=1)
y = df3['positive']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y , test_size=0.3, random_state=42)
```

```python
from sklearn import tree

clf = tree.DecisionTreeClassifier(max_depth=5)
clf= clf.fit(X_train, y_train)
```

```python
clf.score(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
y_pred = clf.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
# RECALL

# 122/(122+28)
recall = cm[0][0]/ (cm[0][0] + cm[0][1])
recall  
```

```python
# PRECISION

presisi = cm[0][0] / (cm[0][0] + cm[1][0])
presisi 
```

```python
lis = []
for i in range(1,20,2):
    clf = tree.DecisionTreeClassifier(max_depth=i)
#     clf = neighbors.KNeighborsClassifier(n_neighbors=i)
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    acc = clf.score(X_test, y_test)
    acc2 = clf.score(X_train, y_train)
    cm = confusion_matrix(y_test, y_pred)
    recall = cm[0][0]/ (cm[0][0] + cm[0][1])
    presisi = cm[0][0] / (cm[0][0] + cm[1][0])
    lis.append([i, acc, acc2, recall, presisi])
    
```
