```python
import pandas as pd
```

```python
df = pd.read_csv('data1.csv', encoding = "ISO-8859-1")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df['InvoiceAmount'] = df['Quantity'] * df['UnitPrice']
```

```python
df
```

```python
df.groupby(['CustomerID']).agg({'InvoiceDate':'max', 'InvoiceAmount':'sum'})
```

```python
df2 = df.groupby(['CustomerID']).agg({'InvoiceNo':'nunique', 
                                'InvoiceDate':'max', 
                                'InvoiceAmount':'sum'})
```

```python
df2.reset_index(inplace=True)
df2
```

```python
import datetime as dt

df2['recency'] = df2['InvoiceDate'].apply(lambda x: (dt.date(2012,1,1) - dt.date(int(x.split()[0].split('/')[2]), 
                                                                                int(x.split()[0].split('/')[0]),
                                                                               int(x.split()[0].split('/')[1]))).days)
```

```python
df2
```

```python
df2.drop('InvoiceDate', inplace=True, axis=1)
df2
```

```python
df2.rename(columns={
    'InvoiceNo': "frequency",
    'InvoiceAmount': 'monetary',
    
}, inplace=True)
```

```python
df2 = df2[df2.monetary >= 0]
```

```python
df2.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
fig, axes = plt.subplots(3)
fig.set_size_inches(5,10)
sn.boxplot(data=df2, y='frequency', orient='v' , ax=axes[0])
sn.boxplot(data=df2, y='recency', orient='v' , ax=axes[1])
sn.boxplot(data=df2, y='monetary', orient='v' , ax=axes[2])
```

```python
def remove_outliers(df, columns, n_std):
    for col in columns:
        print('working on column = {}'.format(col))
        
        mean = df[col].mean()
        sd = df[col].std()
        
        df = df[(df[col] <= mean+(n_std* sd))]
        
    return df
```

```python
df3 = remove_outliers(df2,['frequency', 'recency', 'monetary'], 3)
```

```python
fig, axes = plt.subplots(3)
fig.set_size_inches(5,10)
sn.boxplot(data=df3, y='frequency', orient='v' , ax=axes[0])
sn.boxplot(data=df3, y='recency', orient='v' , ax=axes[1])
sn.boxplot(data=df3, y='monetary', orient='v' , ax=axes[2])
```

```python
df3
```

```python
from sklearn.cluster import KMeans
```

```python
X1 = df2[['recency', 'frequency', 'monetary']].iloc[:, :].values
inertia = []
for n in range(2,20):
    algorithm = (KMeans(n_clusters = n, init='k-means++', n_init=10, max_iter=300,
                      tol=0.0001, random_state=111, algorithm='elkan'))
    algorithm.fit(X1)
    
    inertia.append(algorithm.inertia_)
```

```python
import numpy as np
```

```python
plt.figure(1, figsize=(15,6))
plt.plot(np.arange(2,20), inertia, 'o')
plt.plot(np.arange(2,20), inertia, '-', alpha=0.5)
plt.xlabel('number of clusters'),plt.ylabel('inertia')
plt.show()
```

```python
algorithm = (KMeans(n_clusters = 5, init='k-means++', n_init=10, max_iter=300,
                    tol=0.0001, random_state=111, algorithm='elkan'))
algorithm.fit(X1)
labels3 = algorithm.labels_
centroids3 = algorithm.cluster_centers_
```

```python
df2['Labels']= labels3
```

```python
result = df2.groupby('Labels').agg({'CustomerID': "count",
                                   'frequency': ['mean', 'median', 'min', 'max'],
                                   'monetary': ['mean', 'median', 'min', 'max'],
                                   'recency': ['mean', 'median', 'min', 'max']})
result
```

```python
fig, axes = plt.subplots(nrows=2,ncols=2)
fig.set_size_inches(12, 10)
clusterAggregated = pd.DataFrame(df2.groupby("Labels")["CustomerID"].count()).reset_index()
clusterSorted = clusterAggregated.sort_values(by="CustomerID",ascending=False)
sn.barplot(data=clusterSorted,x="Labels", y="CustomerID",orient="v",ax=axes[0][0])
sn.boxplot(data=df2,y="frequency",x="Labels",orient="v",ax=axes[0][1])
sn.boxplot(data=df2,y="recency",x="Labels",orient="v",ax=axes[1][0])
sn.boxplot(data=df2,y="monetary",x="Labels",orient="v",ax=axes[1][1])

axes[0][0].set(xlabel='Labels', ylabel='Count',title="Bar Plot On Count")
axes[0][1].set(xlabel='Labels', ylabel='Frequency',title="Box Plot On Frequency Across Cluster")
axes[1][0].set(xlabel='Labels', ylabel='Recency',title="Box Plot On Recency Across Cluster")
axes[1][1].set(xlabel='Labels', ylabel='Monetary',title="Box Plot On Monetary Across Cluster")
```
